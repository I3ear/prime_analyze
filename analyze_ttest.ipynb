```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
with open('./result.csv', 'r') as f:
    results = pd.read_csv(f)
```

```python
e_items = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'Valence', 'Arousal', 'like']
primes = ['none', 'bpm', 'chord', 'lyrics', 'playcount']
```

## 対応なし・被験者ごと

```python
results = results.sort_values(by=['m_id', 'prime'])
results
```

```python
results.groupby('prime').mean().loc[primes, e_items]
```

```python
results.groupby('prime').var().loc[primes, e_items]
```

```python
for e_item in e_items:
#     print(e_item)
    none_values = results[results['prime'] == 'none'][e_item]
    for prime in primes[1:]: 
#         print(prime)
        prime_values = results[results['prime'] == prime][e_item]
        t, p = stats.ttest_ind(none_values, prime_values)
        if p <= 0.05:
            print(e_item)
            print(prime)
```

## 対応なし・音楽ごと

```python
results_means = results.groupby(['m_id', 'prime']).mean()
results_means
```

```python
results_means = results_means.reset_index(level=[0, 1])
results_means
```

```python
results_means.groupby('prime').mean().loc[primes, e_items]
```

```python
results_means.groupby('prime').var().loc[primes, e_items]
```

```python
for e_item in e_items:
#     print('-----')
#     print(e_item)
    none_values = results_means[results_means['prime'] == 'none'][e_item]
    for prime in primes[1:]:
#         print(prime)
        prime_values = results_means[results_means['prime'] == prime][e_item]
        t, p = stats.ttest_ind(none_values, prime_values)
        if p <= 0.05:
            print(e_item)
            print(prime)
```

## 対応あり・音楽ごと

```python
for e_item in e_items:
#     print('-----')
#     print(e_item)
    none_values = results_means[results_means['prime'] == 'none'][e_item]
    for prime in primes[1:]:
#         print(prime)
        prime_values = results_means[results_means['prime'] == prime][e_item]
        t, p = stats.ttest_rel(none_values, prime_values)
#         print(p)
        if p <= 0.05:
            print(p)
            print(e_item)
            print(prime)
```

```python
results_items = {}
for e_item in e_items:
    result_item = pd.DataFrame()
    for prime in primes:
        tmp = results_means[results_means['prime'] == prime]
        tmp_series = pd.Series(tmp[e_item].values, index=tmp['m_id'].values)
        result_item[prime] = tmp_series
    results_items[e_item] = result_item
```

```python
for e_item in e_items:
    print('\n\n')
    print(e_item)
    result_item = results_items[e_item]
    none_values = result_item['none'].values
    for prime in primes[1:]:
        print('---')
        print(prime)
        prime_values = result_item[prime].values
        absdiff = np.abs(none_values - prime_values)
        t, p = stats.ttest_1samp(absdiff, 0.0)
        print(p)
```
