```python
import pandas as pd
import pickle
```

```python
moods = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'Valence', 'Arousal']
```

```python
with open('./dataset_audio.tsv', 'r', encoding='shift_jis') as f:
    df_moods = pd.read_csv(f, delimiter='\t', index_col='動画ID')
df_moods = df_moods.iloc[:, 3:11]
df_moods.columns = moods
df_moods[['C1', 'C2', 'C3', 'C4', 'C5', 'C6']] -= 3
df_moods
```

```python
with open('./n_ids.pkl', 'rb') as f:
    n_ids = pickle.load(f)
len(n_ids)
```

```python
df_100 = pd.DataFrame()
for i, n_id in enumerate(n_ids):
    m_id = i + 1
    tmp = df_moods.loc[n_id, :]
    tmp.name = m_id
    df_100 = df_100.append(tmp)
df_100 = df_100[moods]
df_100.index.name = 'm_id'
df_100
```

```python
with open('./moods100.pkl', 'wb') as f:
    pickle.dump(df_100, f)
```
