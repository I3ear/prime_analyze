```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
with open('./result.csv', 'r') as f:
    results = pd.read_csv(f)
results = results.sort_values(by=['m_id', 'prime'])
```

```python
e_items = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'Valence', 'Arousal', 'like']
primes = ['none', 'bpm', 'chord', 'lyrics', 'playcount']
```

```python
nones = results[results['prime'] == 'none']
bpms = results[results['prime'] == 'bpm']
chords = results[results['prime'] == 'chord']
lyrics = results[results['prime'] == 'lyrics']
playcounts = results[results['prime'] == 'playcount']
```

```python
nones.mean()[e_items[:-1]]
```

```python
plt.clf()
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.invert_yaxis()

e_num = len(e_items[:-1])
y = np.array(range(e_num))

ax.plot(nones.mean()[e_items[:-1]], y, marker='s', markersize=5, label='無し')
ax.plot(bpms.mean()[e_items[:-1]], y, marker='s', markersize=5, label='テンポ')
ax.plot(chords.mean()[e_items[:-1]], y, marker='s', markersize=5, label='コード')
ax.plot(lyrics.mean()[e_items[:-1]], y, marker='s', markersize=5, label='歌詞')
ax.plot(playcounts.mean()[e_items[:-1]], y, marker='s', markersize=5, label='再生回数')

ax.xaxis.set_ticks_position('top')
ax.set_yticks(range(e_num))
ax.set_yticklabels(e_items[:-1])
ax.set_xticks([-2, -1, 0, 1, 2])
ax.tick_params(direction='out', pad=30)
ax.legend(loc='upper left')
plt.box('off')

# ax.set_xticklabels(['悪', '小悪', '普', '小善', '善', '最善'])
plt.show()
plt.close()
```
