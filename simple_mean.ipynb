## プレ実験のデータ分析

### 50曲を対象に4種のプライミング＋プライミング無しを比較

- chord：コード進行
- lyrics：歌詞
- bpm：BPM
- playcount：再生回数
- none：なし

```python
import pandas as pd
import numpy as np
# pd.options.display.max_rows = None
```

```python
with open('./result.csv', 'r') as f:
    results = pd.read_csv(f)
```

### 50曲のプライミングパターンごとの評価値（ユーザの平均をとったもの）

```python
results = results.sort_values(by=['m_id', 'prime'])
results = results.groupby(['m_id', 'prime']).mean()
results
```

```python
none = results.xs('none', level=1)
chord = results.xs('chord', level=1)
lyrics = results.xs('lyrics', level=1)
bpm = results.xs('bpm', level=1)
playcount = results.xs('playcount', level=1)
```

```python
len(none[none['like'] == 2])
```

```python
dif_chord = chord - none
# dif_chord
```

```python
dif_lyrics = lyrics - none
# dif_lyrics
```

```python
dif_bpm = bpm - none
# dif_bpm
```

```python
dif_playcount = playcount - none
# dif_playcount
```

### 各プライミングごとに，プライミングなしとの差の（50曲の）平均を求めたもの

```python
output = pd.concat(
    [dif_chord.mean(), dif_lyrics.mean(), dif_bpm.mean(), dif_playcount.mean()],
    axis=1,
    keys=['chord', 'lyrics', 'bpm', 'playcount']
)
output = output.T
output
```

### 結果

- 好感度（like）の中では，lyricsが一番差がでた．ただし平均0.15の差だからかなり低い
- 好感度の平均はプライミングによって下がるものはなかった

```python
label_mapping = {
    label: idx for idx, label in enumerate(np.unique(difs['label']))
}
label_mapping
```
